```python
import requests
from bs4 import BeautifulSoup
from csv import writer
from urllib.request import urlopen
import pandas as pd
import csv
```

```python
response = requests.get("https://www.dell.com/community/Laptops/ct-p/Laptops")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Q1

prod_list = {}
for i in range(5):
        classname = soup.findAll('div', attrs={'class':'cat-card-title'})
        classname = classname[i].text

        post_count = soup.findAll('span', attrs={'class':'cat-card-posts'})
        post_count = post_count[i].text.strip()

        prod_list[classname]=post_count

print("Inspiron post count: ", prod_list["Inspiron"])
print("XPS post count: ", prod_list["XPS"])
print("Latitude post count: ", prod_list["Latitude"])
```

```python
# Q2


file = csv.writer(open('dell_community.csv', 'w', encoding="utf-8"))
file.writerow(['title','author_name','local_time','local_date','kudo count','view count','reply count'])

for j in range(5):
    response = urlopen('https://www.dell.com/community/Inspiron/bd-p/Inspiron/page/'+ str(j+1))
    soup=BeautifulSoup(response,'html.parser')

    all_container = soup.findAll('div',{'class':'lia-component-messages-column-thread-info'})
    all_general2 = soup.findAll('div',{'class':'lia-component-messages-column-message-kudos-count'})
    all_general2_view = soup.findAll('div',{'class':'lia-component-messages-column-topic-views-count'})
    all_general2_replies = soup.findAll('div',{'class':'lia-component-messages-column-message-replies-count'})
    for i in range(len(all_container)):
        g_container=all_container[i]
        gen_container=all_general2[i]
        gen_container_v = all_general2_view[i]
        gen_container_r = all_general2_replies[i]

        # extracting title
        title = g_container.contents[1].a.text.strip()
        #extracting author
        author_name = g_container.contents[3].span.span.a.span.text
        #extracting date/time
        local_time = g_container.findAll('span',{'class':'local-time'})
        local_timef = local_time[0].text
        local_date = g_container.findAll('span',{'class':'local-date'})
        local_datef = local_date[0].text

        #kudo count
        kudo_count = gen_container.findAll('span',{'class','lia-message-stats-count'})
        kudo_count = kudo_count[0].text.strip()

        #view count
        view_count = gen_container_v.findAll('span',{'class','lia-message-stats-count'})
        view_count_final = view_count[0].text

        #count replies
        replies_count= all_general2_replies[i].text.strip()

        print("discussion title"+" "+ title)
        print("author_name"+" "+author_name)
        print("local_time"+" "+local_timef)
        print("local_date"+" "+local_datef)
        print("kudo count"+" "+kudo_count)
        print("view_count"+" "+view_count_final)
        print("reply count"+" "+replies_count)
        print()

    file.writerow([title, author_name,local_timef,local_datef,kudo_count,view_count_final,replies_count])
```
